 # 1.读写文件前先了解一下怎么打开文件
 ### open 函数

- 作用：1.打开一个已经存在的文件
        2.创建一个新文件
- open(文件名，访问模式)
- 有则打开,无则创建。

- ### ***相对路径***

```python
f=open('cheliang.csv','r',encoding='utf-8')
#  r 只读  w 写 x 创建新的
```

```python
# 类型
type(f)
```

```python
# 模式
f.mode
```

```python
f.name #文件的名称
```

```python
#读取f的数据
data=f.read() 
```

```python
# 打印出的内容比较规整
print(data)
```

```python
# 将文件从内存中清除
f.close()
```

 ### ***3.写文件***

- 利用o
pen在桌面创建一个文件

```python
f4=open('C:\\Users\\Administrator\\Desktop\\text.txt','w',encoding='utf-8')
# w 写的模式
```

```python
f4.mode
```

```python
type(f4)
```

```python
# 准备输入的内容
text='''浣溪沙*谁念西风独自凉
       纳兰性德
谁念西风独自凉？
萧萧黄叶闭疏窗。
沉思往事立残阳。
被就莫惊春睡重，赌书消得泼茶香。
当时知道是寻常。'''
```

```python
# 将text的内容写进f4中
f4.write(text)
```

```python
# 若此时打开桌面上的text文件，内容是为空的。
# 需要将f4从内存中清除后，写的内容才会存在于文件中
f4.close()
```
