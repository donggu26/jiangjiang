```python
tips=sns.load_dataset('tips')
tips.head()
```

- ***1、绘制关系图 relplot***

```python
sns.relplot(x='total_bill',y='tip',data=tips);
```

```python
sns.relplot(x='total_bill',y='tip',hue='size',data=tips);
```

```python
sns.scatterplot(x="total_bill", y="tip",hue='size',data=tips);
# 与上述相同的原因是 kind默认的为scatter 
```

```python
sns.scatterplot(x="total_bill", y="tip",hue="day",data=tips,
                size='size', #图形的尺寸
               style='time'); #表示显示的形状依据时间不同显示不同的形状
```

输出多个图形

```python
sns.relplot(x='total_bill', y='tip', hue='time', size='size',
            palette=['g', 'r'],
            sizes=(10, 100),
            col='time',row='sex', # sex与time 进行组合得到4个图形
            data=tips)
```

- ***2、线图 lineplot***

```python
sns.lineplot(x='total_bill',y='tip',data=tips,marker='^');
```

```python
frmi=sns.load_dataset('fmri')
frmi.head()
```

```python
sns.lineplot(x='timepoint', y='signal', data=frmi);
# 产生的阴影是默认的置信区间 默认的是ci=95
```

```python
sns.relplot(x='timepoint',y='signal',data=frmi,
            kind='line',
            marker='*',
            style='region',
           markers=True,
           palette='cool');
```

```python
# 将event设置为分组依据
sns.lineplot(x='timepoint',y='signal',
             hue='event',data=frmi,
             style='event', #按照时间设置形状
             markers=True,
            palette='autumn'); 
```

```python
sns.lineplot(x='timepoint',y='signal',hue='region',data=frmi,
             ci=0, #ci=0表示不设置置信区间
             style='region',markers=True,palette='cool'); #markers=True表示不同额线显示不同的数值的形状
```

# 四、分布图
- 直方图 distplot
- 和密度图  kdeplot
- 变量关系  pairplot
    - 散点矩阵图 scatter_matrix()
       - pd.scatter_matrix() matplotlib库绘制散点矩阵图所用的函数 

```python
sns.distplot(tips['total_bill'],color='purple',axlabel='概率直方图');
# bins 空值柱子的个数
# 默认的是显示直方图和概率密度图
# axlabel : string, False, or None, 设置标签
```

```python
sns.distplot(tips['tip'],kde=False,color='brown',axlabel='直方图');
```

```python
sns.distplot(tips['total_bill'],kde=True,hist=False,color='yellow',axlabel='概率估计图');
```

```python
tips.head()
```

```python
# 双变量的核密度图 
sns.kdeplot(tips['total_bill'],tips['tip'],
            legend=True,color='brown',cbar=True); #cbar 颜色条 显示
```

```python
sns.kdeplot(tips['tip'],legend=True,shade=True,color='red');
```

```python
sns.kdeplot(tips['total_bill'],color='green',
            legend=True, #添加图例
            cumulative=True); #累计
```

# 五、回归图 Regression plots
- lmplot 回归图
- regplot()

```python
sns.lmplot(x='tip',y='total_bill',hue='sex',data=tips,palette='Set2');
```

```python
# 多个图形
# col_wrap=1 表示列上只有一个图形
# height=4，表示有4个图形
# 默认的是带阴影即为95%置信区间 ，可通过x_ci='ci'、ci=95进行调整
sns.lmplot(x="total_bill", y="tip", 
           col="day",hue="day", #列为day 分组依据为day
           data=tips,
           col_wrap=1, height=4);
```

```python
sns.lmplot(x='tip',y='total_bill',data=tips,col_wrap=1,
           hue='sex',col='sex');
# 按照性别为分组依据好的tip与total_bill的回归图
```

```python
sns.lmplot(x='tip',y='total_bill',hue='day',data=tips,markers='*');
```

regplot()

```python
sns.regplot(x='tip',y='total_bill',data=tips,ci=0,color='brown',marker='*');
```

```python
sns.regplot(x='size',y='tip',data=tips,color='purple',marker='^');
```
